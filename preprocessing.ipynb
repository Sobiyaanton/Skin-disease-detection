```python
from PIL import Image, ImageEnhance
import numpy as np
import os
import matplotlib.pyplot as plt

input_folder = '/content/drive/MyDrive/Colab Notebooks/Resized Folder1'
output_folder1 = '/content/drive/MyDrive/Colab Notebooks/contrast1'
output_folder2 = '/content/drive/MyDrive/Colab Notebooks/Psoriasis1'

contrast_levels = [1.5]
enhanced_images = []
for filename in os.listdir(input_folder):
    if filename.lower().endswith(".jpg"):
        img = Image.open(os.path.join(input_folder, filename))
        for level in contrast_levels:
            enhanced_img = ImageEnhance.Contrast(img).enhance(level)
            enhanced_arr = np.asarray(enhanced_img)
            enhanced_images.append(enhanced_arr)
            
            output_filename = f"{os.path.splitext(filename)[0]}.jpg"
            output_path1 = os.path.join(output_folder1, output_filename)
            output_path2 = os.path.join(output_folder2, output_filename)
            Image.fromarray(enhanced_arr).save(output_path1)
            Image.fromarray(enhanced_arr).save(output_path2)
           
plt.figure(figsize=(9,9))
for i in range(min(len(enhanced_images), 16)):
    plt.subplot(4, 4, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(enhanced_images[i])    
plt.show()
```

```python
import cv2
import os
import numpy as np
from skimage.metrics import structural_similarity as ssim

def calculate_ssim(img1, img2):
    ssim_score = ssim(img1, img2, multichannel=True)
    return ssim_score

def compare_images(folder1, folder2):
    ssim_total = 0
    num_images = 0
    
    for filename1 in os.listdir(folder1):
        if not filename1.endswith('.jpg'):
            continue
        
        filename2 = filename1.replace('.jpg', '.jpg')
        filepath1 = os.path.join(folder1, filename1)
        filepath2 = os.path.join(folder2, filename2)
        
        if not os.path.exists(filepath2):
            continue
        
        img1 = cv2.imread(filepath1)
        img2 = cv2.imread(filepath2)
        
        ssim_score = calculate_ssim(img1, img2)
        
        ssim_total += ssim_score
        num_images += 1
        
    ssim_avg = ssim_total / num_images
    
    return ssim_avg

folder1 = "/content/drive/MyDrive/Colab Notebooks/Resized Folder1"
folder2 = "/content/drive/MyDrive/Colab Notebooks/contrast1"

ssim = compare_images(folder1, folder2)

print('Average SSIM:', ssim)
```

```python
import cv2
import os
import numpy as np

def calculate_psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return float('inf')
    PIXEL_MAX = 255.0
    return 20 * np.log10(PIXEL_MAX / np.sqrt(mse))

def compare_images(folder1, folder2):
    psnr_total = 0
    num_images = 0
    
    for filename1 in os.listdir(folder1):
        if not filename1.endswith('.jpg'):
            continue
        
        filename2 = filename1.replace('.jpg', '.jpg')
        filepath1 = os.path.join(folder1, filename1)
        filepath2 = os.path.join(folder2, filename2)
        
        if not os.path.exists(filepath2):
            continue
        
        img1 = cv2.imread(filepath1)
        img2 = cv2.imread(filepath2)
        
        psnr = calculate_psnr(img1, img2)
        
        psnr_total += psnr
        num_images += 1
        
    psnr_avg = psnr_total / num_images
    
    return psnr_avg

folder1 = "/content/drive/MyDrive/Colab Notebooks/Resized Folder1"
folder2 = "/content/drive/MyDrive/Colab Notebooks/contrast1"

psnr = compare_images(folder1, folder2)

print('Average PSNR:', psnr)
```
